```python
from matplotlib import pyplot as plt

eps_3e5 = [0.0268840, 0.0136136, 0.0107648, 0.0092318]
eps_5e6 = [0.0839926, 0.07308984, 0.0492395, 0.0358244]
eps_15e6 = [0.2042897, 0.1403447, 0.1239581, 0.1339453]

num_procs = [1, 4, 16, 32]

plt.figure(figsize=(8,8))
plt.plot(num_procs, eps_3e5, '^-',label="eps=3e-5",)
plt.plot(num_procs, eps_5e6, '^-',label="eps=5e-6")
plt.plot(num_procs, eps_15e6, '^-',label="eps=1.5e-6")
plt.xlabel("numbers of processors")
plt.ylabel("running time")
plt.title("task2 running time comparison")
plt.legend()
plt.grid()
```
